```python
import pandas as pd
```

```python
data=pd.read_csv(r"D:\Internship\D1\netflix_titles.csv")
```

```python
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
#there are more number of director missing so we can drop teh director column
data['director']=data['director'].fillna('unknown')
```

```python
data.isnull().sum()
```

```python
#there are only few DATE_ADDED, ratings, duration values missing so we can drop the rows of that
data=data.dropna(subset=['date_added'])
data
```

```python
data.isnull().sum()
```

```python
# as cast and country are categorical and have fewer missing values we add the missing values with mode values
data['cast']=data['cast'].fillna(data['cast'].mode()[0])
data['country']=data['country'].fillna(data['country'].mode()[0])
```

```python
data.isnull().sum()
```

```python
# as there are only few rows missing in duration and rating we can drop those columns
data=data.dropna(subset=['rating','duration'])
```

```python
data.isna().sum()
```

```python
data
```

```python

data['date_added']=pd.to_datetime(data['date_added'].str.strip(), format="%B %d, %Y")
```
